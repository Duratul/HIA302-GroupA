```python
'DATA PREPARATION'
```

```python
#select columns from cases_malaysia dataset which are "date" and "cases_new" to create a cleaned dataset 'cleaned_df'.
cleaned_df = df_malaysia
cleaned_df = cleaned_df[["date", "cases_new"]]
#convert the "date" to datetime64.
cleaned_df["date"] = pd.to_datetime(cleaned_df["date"],format = "%Y-%m-%d")
#filter the data between 2 dates (from 2022-08-01 until 2022-11-30)
cleaned_df =cleaned_df.loc[(cleaned_df["date"] >= "2022-08-01")
                   & (cleaned_df["date"] < "2022-12-01")]
print(cleaned_df)
```

```python
#review data format
print("Cleaned cases_malaysia dataset dimensions : {}".format(cleaned_df.shape))
```

```python
#review data types, shape, lengths
cleaned_df.info()
```

```python
#checking for missing values, gaps/ lacking data columns using pd.isnull() function.
cleaned_df.isnull().sum()
```

```python
#to show the descriptive analysis of the cleaned dataset.
print ('Descriptive analysis of the cleaned cases_malaysia dataset: ')
print (cleaned_df.describe())
```

```python
'DATA TRANSFORMATION'
```

```python
#to remove duplicates for 'date' column from collected_covid_daily_case dataset for plotting graph visualisation.
collected_remove_dups = df_collected.drop_duplicates('date')
print(collected_remove_dups)
```

```python
#to select 'date', 'holiday' and 'malaysia_cases_new' columns for plotting graph visualisation.
collected_selectdf = collected_remove_dups.iloc[:, [0,1,2]]
print(collected_selectdf)
```

```python
#review data types, shape, lengths
collected_selectdf.info()
```

```python
#checking for missing values, gaps/ lacking data columns using pd.isnull() function.
collected_selectdf.isnull().sum()
```

```python
#to show the descriptive analysis of the common dataset which has cleaned to the purpose of my objective.
print ('Descriptive analysis of the cleaned collected_covid_daily_case dataset dataset: ')
print (collected_selectdf.describe())
```

```python
'DATA VISUALISATION'
```

```python
#import required library
#use altair to plot graph visualisation for cleaned_df dataset
import altair as alt

alt.Chart(cleaned_df).mark_bar().encode(
    x = 'date',
    y = 'cases_new',
    #tooltip is to display x and y variables associated "date" and "cases_new" respectively, and also 'holiday' to show the differences.
    tooltip=['date','cases_new']
    
).properties(width=900)
```

```python
#import required library
#use altair to plot graph visualisation for collected_covid_daily_cases with 'date', 'holiday' and 'malaysia_cases_new' columns.

alt.Chart(collected_selectdf).mark_bar().encode(
    x = 'date',
    y = 'malaysia_cases_new',
    #tooltip is to display x and y variables associated 'date' and 'malaysia_cases_new' respectively, and also 'holiday' to show the differences.
    tooltip=['date','holiday','malaysia_cases_new'],
    color='holiday'
    
).properties(width=800)
```
